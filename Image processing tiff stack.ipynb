Grainy images can be a real pain to process. In this guide I will demonstrate how to process a grainy image using Scikit (there are equivalent filters in Matlab) and generate intensity curves from the bubble. So far I have only used this code on TIRF bubbles. 

**Requirements**. I used the scikit-image library for the image processing routines. To handle TIFF stacks I used the PIMS library (https://github.com/soft-matter/pims) --  I have had no luck with other image libraries for Python.

The processing sequence is as follows. 

1. Apply a median filter. For TIRF images 1-2 pixels are enough, for TEM it will be more. 
2. Equalise histogram (default settings).
3. Apply a Canny filter (the default setting is $\sigma$=1, it's usually not wise to go far beyond this value).
4. Apply an entropy filter with radius 2.
5. Use a morphological closing to remove excess specks. Clearly this will only work if your specks are smaller than the bubble you're imaging. Here the bubble is $\sim$12 pixels so I morphologically closed all features smaller than 10 pixels.

```python
from skimage import exposure, img_as_bool
from skimage.filter.rank import median, entropy
from skimage.morphology import disk, erosion, dilation, opening, closing, white_tophat

# Image parameters
img = v[2]
pixelWidthJ = img.shape[0]
pixelWidthI = img.shape[1]

# Step 1: 1-2 pixel median filter
median_v = median(img, disk(1))

# Equalise histogram
equalise_v = exposure.equalize_hist(median_v)

# Canny filter, default settings
edges = filter.canny(equalise_v)
plt.subplot(221)
plt.imshow(edges, cmap = cm.gray)
plt.title('Canny filter')

# Entropy filter
entropy_v = entropy(edges,disk(2))
plt.subplot(222)
plt.title('Entropy filter')
plt.imshow(entropy_v, cmap=cm.gray)

bool_v = img_as_bool(entropy_v)
plt.subplot(223)
plt.title('Boolean entropy')
plt.imshow(bool_v, cmap = cm.gray)

# Finally, morphological closing
selem = disk(10)  # All features smaller than this will be ignored
closed_v = closing(bool_v, selem)
plt.subplot(224)
plt.title('Morph. close')
plt.imshow(closed_v, cmap = cm.gray)
plt.tight_layout()
```

The morphologically-closed region provides a binary mask on which we can calculate pixel intensities.

## The intensity code in a function

```python
fileName = 'tap-fv on b5-medium.tif'

def imageIntensity(fileName, frameNum): 
    
    # Image parameters
    v = pims.TiffStack(fileName, as_grey=True)
    img = v[frameNum]
    pixelWidthJ = img.shape[0]
    pixelWidthI = img.shape[1]

    # Step 1: 1-2 pixel median filter
    median_v = median(img, disk(1))

    # Equalise histogram
    equalise_v = exposure.equalize_hist(median_v)

    # Canny filter, default settings
    edges = filter.canny(equalise_v)

    # Entropy filter
    entropy_v = entropy(edges,disk(2))
    bool_v = img_as_bool(entropy_v)

    # Finally, morphological closing
    selem = disk(10)  # All features smaller than this will be ignored
    closed_v = closing(bool_v, selem)
    
    # Calculate the output intensity
    outputIntensity = 0
    outputArea = 0
    for i in range(pixelWidthI):
        for j in range(pixelWidthJ):
            if closed_v[i,j] == 0:
                outputIntensity += img[i,j]
                outputArea += 1
                
    return outputIntensity, outputArea
```

We can now plot the intensities for the frames in the video. I've also plotted the intensity averaged by the bubble area. We get an intensity that peaks every 10 frames or so. 

```python
startFrame = 0
endFrame = 50
fileName = 'tap-fv on b5-medium.tif'

x = []
y = []
z = []

for i in range(endFrame-startFrame):
    currValue, currIntensity = imageIntensity(fileName,i)
    if currValue > 2e5: 
        y = append(y,currValue)
        z = append(z, currValue/currIntensity)
        x = append(x,i)
        
plt.figure(figsize=(10,5))
plt.subplot(121)
plt.plot(x,y)
plt.xlabel('Frame number')
plt.ylabel('Intensity')

plt.subplot(122)
plt.plot(x,z)
plt.xlabel('Frame number')
plt.ylabel('Average intensity')
plt.tight_layout()
```
